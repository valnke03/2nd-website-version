```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from nilearn import datasets

# --- 1. FETCH DATA ---
# Fetch 50 subjects to ensure there is enough for a good group comparison.
# pipeline='cpac': Standard preprocessing.
# derivatives=['rois_aal']: Timeseries extracted using AAL atlas (116 regions).
print("Fetching ABIDE dataset...")
abide = datasets.fetch_abide_pcp(n_subjects=50, pipeline='cpac', derivatives=['rois_aal'])

# --- 2. PREPARE & VISUALIZE ONE SUBJECT (FIXED) ---
subject_id = 0

# Use .iloc to access the label by ROW POSITION, not by index ID
label_val = abide.phenotypic['DX_GROUP'].iloc[subject_id]
label_text = 'Autism' if label_val == 1 else 'Control'

# Get the time series for this subject
time_series = abide.rois_aal[subject_id]

# Compute Correlation (Adjacency) Matrix
correlation_matrix = np.corrcoef(time_series.T)

print(f"\nSubject Index: {subject_id}")
print(f"Classification: {label_text}")
print(f"Data Shape: {time_series.shape} (Time Points x Regions)")

# Plot Subject 0
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm', center=0, vmin=-1, vmax=1)
plt.title(f"Subject {subject_id} Connectivity Graph\nClass: {label_text}")
plt.xlabel("Brain Regions (1-116)")
plt.ylabel("Brain Regions (1-116)")
plt.show()

# --- 3. GROUP ANALYSIS (FEATURE ENGINEERING) ---
print("\n--- Starting Group Analysis ---")
autism_matrices = []
control_matrices = []

# Loop through all downloaded subjects
for i in range(len(abide.rois_aal)):
    # Get Label (1=Autism, 2=Control)
    group_label = abide.phenotypic['DX_GROUP'].iloc[i]

    # Get Data & Compute Correlation
    ts = abide.rois_aal[i]
    corr_mat = np.corrcoef(ts.T)

    # Sort into lists
    if group_label == 1:
        autism_matrices.append(corr_mat)
    elif group_label == 2:
        control_matrices.append(corr_mat)

print(f"Total Subjects: {len(abide.rois_aal)}")
print(f"Autism Group: {len(autism_matrices)} subjects")
print(f"Control Group: {len(control_matrices)} subjects")

# Compute Average Brains
if len(autism_matrices) > 0 and len(control_matrices) > 0:
    avg_autism = np.mean(autism_matrices, axis=0)
    avg_control = np.mean(control_matrices, axis=0)

    # Visualize the DIFFERENCE (The "Signal" your model needs to learn)
    difference_matrix = avg_autism - avg_control

    plt.figure(figsize=(10, 8))
    sns.heatmap(difference_matrix, cmap='coolwarm', center=0)
    plt.title("Mean Difference (Autism - Control)\nRed = Stronger in Autism, Blue = Stronger in Control")
    plt.xlabel("Brain Regions")
    plt.ylabel("Brain Regions")
    plt.show()
else:
    print("Error: One of the groups is empty. Try increasing n_subjects in step 1.")
```
